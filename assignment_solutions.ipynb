
# Python Basics Theory Answers

1. **What is Python, and why is it popular?**
Python is a high-level, interpreted programming language known for its readability, versatility, and large community support. It is popular due to its simplicity, vast libraries, and applications in AI, ML, web development, and automation.

2. **What is an interpreter in Python?**
An interpreter executes code line by line, converting source code into machine code at runtime.

3. **What are pre-defined keywords in Python?**
Keywords are reserved words that have special meaning, such as `if`, `for`, `while`, `class`.

4. **Can keywords be used as variable names?**
No, keywords cannot be used as identifiers.

5. **What is mutability in Python?**
Mutability refers to whether an objectâ€™s value can be changed after creation.

6. **Why are lists mutable, but tuples are immutable?**
Lists store references that allow item modification, while tuples store fixed references preventing changes.

7. **Difference between `==` and `is`:**
`==` checks value equality, `is` checks identity (memory location).

8. **Logical operators:**
`and`, `or`, `not`.

9. **Type casting:**
Converting data type of a value.

10. **Implicit vs explicit casting:**
Implicit is automatic conversion, explicit is manually done using functions.

11. **Purpose of conditional statements:**
To control flow of program based on conditions.

12. **elif statement:**
Used for multiple condition checks.

13. **for vs while:**
`for` is used with iterable-based loops, `while` runs until condition false.

14. **Scenario for while loop:**
When number of iterations isn't known beforehand, e.g., reading user input until they enter quit.

```
print(sum(i for i in range(1,51) if i%2==0))
```

### Reverse string

```
s='hello'; rev=''; i=len(s)-1
while i>=0: rev+=s[i]; i-=1
print(rev)
```

### Factorial

```
n=5; f=1; i=1
while i<=n: f*=i; i+=1
print(f)
```
