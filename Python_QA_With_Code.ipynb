### Q1. Create a string with your name and print it

```python
name = 'Mohammed Tabran'
print(name)
```

### Q2. Find the length of the string "Hello World"
The `len()` function returns the number of characters in a string.

```python
text = 'Hello World'
print(len(text))
```

### Q3. Slice first 3 characters from "Python Programming"
Slicing extracts a part of a string.

```python
text = 'Python Programming'
print(text[:3])
```

```python
print('hello'.upper())
```
